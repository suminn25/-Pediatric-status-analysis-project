```python
df1_21 = df1.loc[df1['기준분기'] == 202104]
df1_21
```

```python
df1_21.reset_index(drop=True, inplace=True)
df1_21 = df1_21.set_index('시도')
del df1_21['기준분기']
df1_21.head()
```

```python
df1_21.info()
```

```python
df1_21 = df1_21[['소아청소년과']]
```

```python
df1_21
```

```python
df1_21 = df1_21.drop(index = '전국')
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'Malgun Gothic'
```

```python
data = df1_21['소아청소년과'].value_counts().sort_index(ascending = False)
```

```python
plt.xticks(rotation=45)
plt.ylabel('의원 수')
plt.xlabel('지역')
plt.title('지역별 소아청소년과 의원 수')
plt.bar(df1_21.index, df1_21['소아청소년과'])
# plt.legend([''])
```

```python
df2_21 = df2[['2021']]
df2_21
```

```python
df2_21 = df2_21.drop(index = '전국')
```

```python
df2_21
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 4))
plt.ylabel('출생아 수')
plt.xlabel('지역')
plt.title('21년도 지역별 출생아 수')
plt.bar(df2_21.index, df2_21['2021'])
# plt.legend([''])
```

```python
total_df21 = pd.merge(df1_21, df2_21, left_on = '시도', right_on = '시도', how = 'inner')
total_df21
```

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax = sns.scatterplot(data=total_df21, x='소아청소년과', y='2021')
plt.ylabel('출생아 수')
plt.xlabel('소아청소년과')
plt.title('지역별 의원 수와 출생아 수의 관계')
plt.legend(['지역'])
plt.show()
```
