```python
df2 = df2[['시도', '2021', '2020', '2019', '2018']]
df2
```

```python
total = pd.merge(df1, df2, left_on = '지역', right_on = '시도', how = 'inner')
total
```

```python
del total['시도']
```

```python
t_21 = total['2021_y'] / total['2021_x']
t_21
```
